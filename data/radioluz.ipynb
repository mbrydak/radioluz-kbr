```python
import pandas as pd
from pandas_ods_reader import read_ods
import os
```

```python
ods_files = []
excel_files = []
for filename in os.scandir("data_raw"):
    if filename.is_file():
        if filename.name.__contains__('.xls') or filename.name.__contains__('.xlsx'):
            excel_files.append(filename.path)
        elif filename.name.__contains__('.ods'):
            ods_files.append(filename.path)
```
